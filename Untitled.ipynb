```python
df=sns.load_dataset("iris")
```

```python
setosa_val=df.sepal_width[df.species=="setosa"].value_counts().values
```

```python
setosa_width=df.sepal_width[df.species=="setosa"].value_counts().index
```

```python
sns.displot(df.sepal_width[df.species=="setosa"],kde=True)
```

```python
sns.relplot(data=df,x="petal_width",y="sepal_length",hue="species")
plt.title("Sepal Length VS Petal Width")
```

```python
sns.relplot(data=df,x="sepal_length",y="sepal_width",hue="species")
plt.title("Sepal Length VS Sepal Width")
```

```python
sns.relplot(data=df,x="petal_length",y="petal_width",hue="species")
plt.title("Petal Length VS Petal Width")
```

```python
sns.relplot(data=df,x="petal_length",y="petal_width",hue="species",kind="line")
plt.title("Petal Length VS Petal Width")
```

```python
sns.catplot(data=df,x="species",y="sepal_length")
```

```python
sns.boxplot(data=df,x="species",y="petal_length")
```

```python
sns.boxenplot(data=df,x="species",y="sepal_length")
```

```python
sns.swarmplot(data=df,x="species",y="petal_width")
```

```python
sns.regplot(data=df,x="petal_length",y="sepal_length",marker="X")
```

```python
sns.regplot(data=df,x="petal_length",y="petal_width",marker="X")
```
